```python
import torch
```

```python
pool = torch.nn.AvgPool2d(16)
mean_val = 0.6
```

```python
x = torch.randn((1, 3, 256, 256))
```

```python
x.shape
```

```python
x_mean = torch.mean(x, 1, keepdim=True)
x_mean.shape
```

```python
x_pool = pool(x_mean)
x_pool.shape
```

```python
out = torch.pow(x_pool - torch.FloatTensor([mean_val]), 2)
```

```python
out.shape
```

```python
out = torch.mean(out)
out.shape
```
